```python
from scipy import array, arange, sqrt,exp
from pylab import plot, show, xlabel, ylabel

xo=1
n=100 #numero de pasos 
x=[]
x.append (xo)
for i in range(100):
    xo= 2-exp(-xo)
    x.append(xo)
    
print (x[len(x)-1])
plot(x,'g')
show()
```

en esta ocasión se puede ver claramente como al solucionar la ecuación, esta converge a un valor mostrando que el método es efectivo para este tipo de ecuación x=f(x), mostrando un comportamiento lineal a su tendencia.

```python
from scipy import array, arange, sqrt,exp,log
from pylab import plot, show, xlabel, ylabel
xo=0.5
n=100 #numero de pasos 
x=[]
x.append (xo)
for i in range(100):
    xo= sqrt(1-log(xo))
    x.append(xo)
    
print (x[len(x)-1])
plot(x,'r')
show()
```
